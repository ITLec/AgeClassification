```python
class_num=1
# from keras.layers import Input
model = tf.keras.Sequential([
    pretrained_model,
    # tf.keras.layers.Dense(256, activation='relu'),
    # tf.keras.layers.Dropout(0.5),
        # tf.keras.layers.Dense(128, activation='relu'),
    # tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(class_num, activation='sigmoid')
])
```

```python
import numpy as np
probabilities = model.predict(X_test_scaled) # Get probabilities
predictions = np.where(probabilities > 0.16161616161616163, 1, 0) # Apply threshold
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y, predictions))
print(classification_report(y, predictions))
```

```python
from sklearn.metrics import precision_recall_fscore_support

# Calculate probabilities
probabilities = model.predict(X_test_scaled)

# Calculate precision, recall, and F1-score for different thresholds
thresholds = np.linspace(0, 1, num=100)  # Generate a range of thresholds
f1_scores = []
for threshold in thresholds:
    predictions = np.where(probabilities > threshold, 1, 0)
    _, _, f1_, _ = precision_recall_fscore_support(y, predictions, average='binary')
    f1_scores.append(f1_)

# Find the threshold with the best F1-score
best_threshold_index = np.argmax(f1_scores)
best_threshold = thresholds[best_threshold_index]
best_f1_score = f1_scores[best_threshold_index]

# Use the best threshold for final predictions
final_predictions = np.where(probabilities > best_threshold, 1, 0)

print("Best Threshold:", best_threshold)
print("Best F1-score:", best_f1_score)
```
